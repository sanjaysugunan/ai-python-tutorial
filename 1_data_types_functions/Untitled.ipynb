```python
revenues = [10,20,30]
expenses = [11,30,55]
```

```python
for rev in revenues:
    print("Revenues: ", rev)
```

```python
for i in range(len(revenues)):
    profit = revenues[i]-expenses[i]
    margin = profit*100/revenues[i]
    print(f"Margin Q{i+1}: {margin:.2f}")
```

```python
margins = []

for i in range(len(revenues)):
    profit = revenues[i]-expenses[i]
    margin = profit*100/revenues[i]
    margins.append(margin)

margins
```

```python
financials = [{"revenue": 50, "expenses": 10}, # q1
              {"revenue": 60, "expenses": 20}, # q2
              {"revenue": 70, "expenses": 30}  # q3
             ]
financials
```

```python
for fin in financials:
    profit = fin["revenue"] - fin["expenses"]
    fin["margin"] = profit * 100 / fin["revenue"]

financials
```

```python
financials = {
    "Q1": {'revenue': 50, 'expenses': 10},
    "Q2": {'revenue': 60, 'expenses': 20},
    "Q3": {'revenue': 70, 'expenses': 30}
}

for quarter, data in financials.items():
    margin = (data["revenue"] - data["expenses"]) / data["revenue"] * 100
    print(f"{quarter}: {margin:.2f}%")
```

```python
financials = {
    "Q1": {'revenue': 50, 'expenses': 10},
    "Q2": {'revenue': 60, 'expenses': 20},
    "Q3": {'revenue': 70, 'expenses': 30},
    "Q4": {'revenue': -10, 'expenses': 35}
}

for quarter, data in financials.items():
    if data["revenue"]<0:
        print(f"Invalid revenue for {quarter}. Skipping.")
        continue
    margin = (data["revenue"] - data["expenses"]) / data["revenue"] * 100
    print(f"{quarter}: {margin:.2f}%")
```
