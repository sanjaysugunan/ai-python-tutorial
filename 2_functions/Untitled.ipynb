```python
def calculate_profit_and_margin(revenue, expenses):
    profit = revenue - expenses
    margin = profit * 100 / revenue
    return profit, margin
```

```python
revenue = 50
expenses = 10

profit, margin = calculate_profit_and_margin(revenue, expenses)
profit, margin
```

```python
## type hint
def calculate_margin(revenue: float, expenses: float) -> float:
    profit = revenue - expenses
    margin = profit * 100 / revenue
    return margin
```

```python
def calc_total_expense(*args):
    return sum(args)

calc_total_expense(10,1,1,23,4,5)
```

```python
def total_expenses(**kargs):
    total = 0
    for key, value in kargs.items():
        total += value
    print(kargs)
    return total

total = total_expenses(rent = 1000, phone_bill = 30, car = 100)
print(f"total: {total}")
```

## class

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def calculate__bonus(self):
        return self.salary * 0.10

emp1 = Employee("Mohan", 80000)

emp1.calculate__bonus()
```

```python
from datetime import date

class Human:

    def __init__(self, name, birth__year):
        self.name = name
        self.birth__year = birth__year

    def get_age(self):
        current__year = date.today().year
        return current__year - self.birth__year

    def introduce(self):
        print(f"Hi, my name is {self.name}. I am {self.get_age()} years old")

human1 = Human("Nigesh", 1998)
human1.introduce()
```
