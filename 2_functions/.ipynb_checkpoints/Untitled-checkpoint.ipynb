```python
def calculate_margin(revenue, expenses):
    profit = revenue - expenses
    margin = profit * 100 / revenue
    return margin
```

```python
calculate_margin(50,10)
```

```python
def calculate_profit_and_margin(revenue, expenses):
    profit = revenue - expenses
    margin = profit * 100 / revenue
    return profit, margin
```

```python
revenue = 50
expenses = 10

profit, margin = calculate_profit_and_margin(revenue, expenses)
profit, margin
```

```python
respone = calculate_profit_and_margin(revenue, expenses)
type(respone)
```

```python
## type hint
def calculate_margin(revenue: float, expenses: float) -> float:
    profit = revenue - expenses
    margin = profit * 100 / revenue
    return margin
```

```python
def calc_total_expense(*args):
    return sum(args)

calc_total_expense(10,1,1,23,4,5)
```

```python
def total_expenses(**kargs):
    print(kargs)

total_expenses(rent = 100, phone_bill = 30, msg = "hi")
```
