```python
try:
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print("Rate:", data["rate"])
    else:
        print("Something went wrong. Status code:", response.status_code)

except Exception as e:
    print("Could not reach the API:", e)
    
```
