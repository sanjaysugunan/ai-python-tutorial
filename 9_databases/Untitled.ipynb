```python
import sqlite3
```

```python
conn = sqlite3.connect("shop.db")

cursor = conn.cursor()

print("database is ready")
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS reviews(
        id INTEGER PRIMARY KEY,
        text TEXT,
        laabel TEXT,
        score INTEGER
    )"""
)
```

```python
cursor.execute(
    "INSERT INTO reviews (text, laabel, score) VALUES (?,?,?)",
    ("The biryani was amazing!","positive", 5)
)
```

```python
conn.commit()
```

```python
more_reviews = [("The biryani was super!","positive", 5),
                ("The biryani was bad!","neutral", 2)]
cursor.executemany(
    "INSERT INTO reviews (text, laabel, score) VALUES (?,?,?)",
    more_reviews
)

conn.commit()
```

```python
cursor.execute("SELECT * from reviews")

rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
conn.close()
```
