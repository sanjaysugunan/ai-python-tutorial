```python
from dotenv import load_dotenv
from google import genai

load_dotenv()
client = genai.Client()
```

```python
response = client.models.generate_content(
        model="gemini-3.5-flash",
        contents = "Give me a simple Indian dal recipe (in less than 10 lines) with calorie count."
)

print(response.text)
```

# for structuring the output

```python
from pydantic import BaseModel
from google.genai import types

# This class is the shape we want back. Note the types: str, int, list of str.
class Recipe(BaseModel):
    title: str
    ingredients: list[str]
    calories: int
    prep_time_minutes: int

response =  client.models.generate_content(
    model = "gemini-2.5-flash",
    contents = "Give me a simple Indian dal recipe",
    config = types.GenerateContentConfig(
        response_mime_type = "application/json",  # reply in JSON
        response_schema =  Recipe,
    ),
)

print(response.text)
```
