```python
from google import genai

client = genai.Client()

interaction = client.interactions.create(
    model="gemini-3.5-flash",
    input="Explain how AI works in a few words"
)

print(interaction.output_text)
```

```python
ingradients = ["dal","onion","chicken"]
'.'.join(ingradients)
```

```python
def generate_recipe(ingredients, cuisine, diet):
    prompt = f'''
    Generate one food recipe using these ingredients: {','.join(ingredients)}
    Recipe should not be more than 100 words
    Cuise: {cuisine}
    Diet: {diet}
    '''
    try:
        response = client.models.generate_content(
        model="gemini-3.5-flash",
        contents = prompt
        )
        if not response.text:
            return "The model returned an empty response. Please try again."
        return response.text
    except Exception as e:
        # Network down, bad API, I ate limit etc, etc.
        return f"Sorry, couldn't generate a recipe right now. ({e})"
    
```

```python
recipe = generate_recipe(["Chicken", "masalas", "butter", "tomatoes"],"Indian","Non-veg")
recipe
```
